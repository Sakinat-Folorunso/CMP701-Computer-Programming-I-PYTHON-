## üß† Short Concept Recap (From Lecture)

Decision making allows a program to **choose between alternative actions**.

In real life:
- If it rains ‚Üí carry umbrella  
- If score ‚â• 50 ‚Üí pass  

In Python, decisions are made using **Boolean expressions** that evaluate to `True` or `False`.



## üß™ Lab 5.1: Boolean Expressions

Boolean expressions return either `True` or `False`.


```python

# Examples of Boolean expressions
print(5 > 3)        # Greater than
print(10 == 10)     # Equality check
print(7 != 2)       # Not equal
print(4 <= 6)       # Less than or equal
```


## üß™ Lab 5.2: The if Statement

The `if` statement executes code **only when a condition is true**.


```python

# Input student score
score = int(input("Enter score: "))

# Check pass condition
if score >= 50:
    print("Passed")
```


## üß™ Lab 5.3: if‚Äìelse Statement

The `if‚Äìelse` statement provides **two alternative paths**.


```python

# Input student score
score = int(input("Enter score: "))

# Determine pass or fail
if score >= 50:
    print("Passed")
else:
    print("Failed")
```


## üß™ Lab 5.4: elif Statement (Multiple Conditions)

Used when **more than two outcomes** are possible.


```python

# Input student score
score = int(input("Enter score: "))

# Determine grade
if score >= 70:
    print("Grade A")
elif score >= 60:
    print("Grade B")
elif score >= 50:
    print("Grade C")
else:
    print("Fail")
```


## üß† Execution Flow (Think Before You Run)

Only **one block** of an if‚Äìelif‚Äìelse structure executes.



## üß™ Lab 5.5: Nested if Statement

An `if` statement inside another `if` statement.


```python

# Input age and income
age = int(input("Enter age: "))
income = float(input("Enter monthly income: "))

# Loan eligibility check
if age >= 21:
    if income >= 50000:
        print("Eligible for loan")
    else:
        print("Income too low")
else:
    print("Age not eligible")
```


## ‚ö†Ô∏è Common Errors to Avoid
- Using `=` instead of `==`
- Incorrect indentation
- Overlapping elif conditions
- Forgetting the final else
