## üß† Short Concept Recap (From Lecture)

In **Week 06**, we used the `while` loop for condition-controlled repetition.

In **Week 07**, we use the `for` loop when:
- The number of repetitions is known
- We are iterating over a sequence (numbers, list, string)

Think of the `for` loop as:
> ‚ÄúFor each value in this sequence, do something.‚Äù



## üîÅ while vs for (Quick Comparison)

| Feature | while loop | for loop |
|------|-----------|---------|
| Control | Condition-based | Count/sequence-based |
| Best for | Unknown repetitions | Known repetitions |
| Variable update | Manual | Automatic |



## üß™ Lab 7.1: Basic for Loop


```

# Print numbers from 1 to 5 using for loop
for i in range(1, 6):
    print(i)
```


### üîç Explanation
- `range(1, 6)` generates numbers from 1 to 5
- The loop variable `i` changes automatically



## üß™ Lab 7.2: Understanding range()

The range() function can take 1, 2, or 3 arguments.


```

# range(stop)
print(list(range(5)))

# range(start, stop)
print(list(range(1, 6)))

# range(start, stop, step)
print(list(range(1, 11, 2)))
```


## üß™ Lab 7.3: Multiplication Table


```

# Input a number
num = int(input("Enter a number: "))

# Generate multiplication table
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```


## üß™ Lab 7.4: break Statement

The break statement stops the loop immediately.


```

# Stop loop when i reaches 5
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```


## üß™ Lab 7.5: continue Statement

The continue statement skips the current iteration.


```

# Skip printing number 3
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```


## üß™ Lab 7.6: Nested Loops

A loop inside another loop.


```

# Print a square pattern using nested loops
for i in range(1, 4):
    for j in range(1, 4):
        print("*", end=" ")
    print()  # New line after inner loop
```


## üß† Dry-Run Thinking

Outer loop controls rows  
Inner loop controls columns
