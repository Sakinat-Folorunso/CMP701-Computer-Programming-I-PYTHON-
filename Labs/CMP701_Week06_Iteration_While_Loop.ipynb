## üß† Short Concept Recap (From Lecture)

Many problems require **repeating actions**:
- Counting numbers
- Summing values
- Repeating input until a condition is met

In Python, repetition is achieved using **loops**.
This week focuses on the **while loop**, which repeats **as long as a condition is true**.



## üîÅ The while Loop Structure

General syntax:
```python
while condition:
    statement(s)
```
The loop continues **while the condition is True**.



## üß™ Lab 6.1: Counting Numbers Using while Loop


````python

# Initialize a counter variable
i = 1

# Loop continues while i is less than or equal to 5
while i <= 5:
    print(i)      # Display current value of i
    i += 1        # Increment i by 1
````


### üîç Explanation
- `i` is the loop control variable
- The condition is checked before each iteration
- The loop stops when the condition becomes False



## üß™ Lab 6.2: Sum of First 10 Natural Numbers


````python

# Initialize sum and counter
total = 0
i = 1

# Loop to compute sum
while i <= 10:
    total += i    # Add current i to total
    i += 1        # Move to next number

# Display result
print("Sum =", total)
````


## üß† Dry-Run Table (Very Important)

For the loop above:
| Iteration | i | total |
|----------|---|-------|
| 1 | 1 | 1 |
| 2 | 2 | 3 |
| ... | ... | ... |



## üß™ Lab 6.3: Password Retry System

This program keeps asking for a password until the correct one is entered.


````python

# Correct password
password = "admin"

# Initial user input
user_input = ""

# Repeat until correct password is entered
while user_input != password:
    user_input = input("Enter password: ")

print("Access granted")
````
