3. Design an algorithm

## üß© Lab 2.1: Understanding a Simple Problem

### Problem Statement  
Find the **maximum of two numbers** entered by the user.

### Step 1: Identify Inputs and Outputs
- Inputs: Two numbers (a, b)
- Output: The larger number

### Step 2: Algorithm (in plain English)
1. Read the first number
2. Read the second number
3. Compare both numbers
4. Display the larger number



## ‚úçÔ∏è Pseudocode Representation

```
START
INPUT a, b
IF a > b THEN
    PRINT a
ELSE
    PRINT b
STOP
```



## üß™ Lab 2.2: Converting Algorithm to Python Code
Below is the Python implementation of the algorithm.
Study each line carefully.


````python

# Ask the user to enter the first number
a = int(input("Enter first number: "))

# Ask the user to enter the second number
b = int(input("Enter second number: "))

# Compare the two numbers
if a > b:
    # If a is greater than b, display a
    print("Maximum number is:", a)
else:
    # Otherwise, display b
    print("Maximum number is:", b)
````


## üîç Code Explanation
- `input()` collects data from the user
- `int()` converts input text to integer
- `if` checks a condition
- Only **one block** runs depending on the condition



## üß™ Lab 2.3: Student Average Score

### Problem Statement  
Compute the **average score of a student** using three test scores.


````python

# Input the three scores from the user
score1 = int(input("Enter score 1: "))
score2 = int(input("Enter score 2: "))
score3 = int(input("Enter score 3: "))

# Calculate the average score
average = (score1 + score2 + score3) / 3

# Display the average score
print("Average score =", average)
````

## üß™ Lab 2.4: Algorithm ‚Üí Code (ATM Withdrawal ‚Äì Simplified)

### Problem Description
Simulate a simple ATM withdrawal system.
- Input: account balance, withdrawal amount
- Condition: withdrawal must not exceed balance


````python

# Input account balance
balance = float(input("Enter account balance: "))

# Input withdrawal amount
amount = float(input("Enter withdrawal amount: "))

# Check if withdrawal is possible
if amount <= balance:
    balance = balance - amount
    print("Withdrawal successful")
    print("Remaining balance =", balance)
else:
    print("Insufficient balance")
````


## ‚ö†Ô∏è Common Mistakes to Avoid
- Skipping algorithm design
- Writing code without understanding the problem
- Forgetting type conversion (`int`, `float`)
- Poor indentation in if-statements
