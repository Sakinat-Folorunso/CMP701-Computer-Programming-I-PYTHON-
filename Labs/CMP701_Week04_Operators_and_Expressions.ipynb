## ðŸ§  Short Concept Recap (From Lecture)

An **operator** tells Python to perform an operation.  
An **expression** is a combination of operators and operands that produces a value.

Example:
```
5 + 3 * 2
```
This is an expression containing operators `+` and `*`.



## ðŸ§ª Lab 4.1: Arithmetic Operators

Arithmetic operators are used for mathematical calculations.


````python

# Addition
print(5 + 3)

# Subtraction
print(10 - 4)

# Multiplication
print(6 * 7)

# Division (always returns float)
print(10 / 4)

# Modulus (remainder)
print(10 % 4)

# Exponentiation
print(2 ** 3)

# Floor division
print(10 // 4)
````


### ðŸ” Explanation
- `/` gives decimal result
- `//` removes decimal part
- `%` gives the remainder



## ðŸ§ª Lab 4.2: Relational (Comparison) Operators

Relational operators compare values and return **True** or **False**.


````python

# Greater than
print(10 > 5)

# Less than
print(3 < 2)

# Equal to
print(5 == 5)

# Not equal to
print(5 != 3)

# Greater than or equal to
print(7 >= 7)

# Less than or equal to
print(4 <= 6)
````


## ðŸ§ª Lab 4.3: Logical Operators

Logical operators combine multiple conditions.


````python

# Logical AND
print(True and False)

# Logical OR
print(True or False)

# Logical NOT
print(not True)
````


## ðŸ§  Truth Table Reminder

## ðŸ§ª Lab 4.4: Operator Precedence

Python evaluates expressions based on precedence rules.


````python

# Operator precedence example
result = 10 + 5 * 2
print(result)

# Using parentheses to change order
result = (10 + 5) * 2
print(result)
````


### ðŸ§  Dry-Run Explanation
- Multiplication happens before addition
- Parentheses override precedence



## ðŸ§ª Lab 4.5: Real-World Example â€“ Salary Computation


````python

# Input basic salary
basic_salary = float(input("Enter basic salary: "))

# Compute allowance (20% of basic salary)
allowance = 0.2 * basic_salary

# Compute gross salary
gross_salary = basic_salary + allowance

# Display result
print("Gross Salary =", gross_salary)
````
