## üß† Concept Recap: From Lists to Records

Lists store values by position.
However, real-world data often has **meaningful labels**.

Example:
Student record ‚Üí name, matric, department, score

This motivates **dictionaries**.



## üß™ Lab 9.1: Creating a Dictionary

A dictionary stores data as **key : value** pairs.


```

# Create a dictionary for a student
student = {
    "name": "Aisha",
    "matric": "CSC/PGD/001",
    "department": "Computer Science",
    "score": 78
}

# Display dictionary
print(student)
```


## üß™ Lab 9.2: Accessing Dictionary Values


```

# Access values using keys
print(student["name"])
print(student.get("score"))  # safer access
```


## üß™ Lab 9.3: Updating and Adding Items


```

# Update score
student["score"] = 85

# Add new key-value pair
student["level"] = "PGD"

print(student)
```


## üß™ Lab 9.4: Removing Items


```

# Remove department
student.pop("department")

print(student)
```


## üß™ Lab 9.5: Looping Through Dictionaries


```

# Loop through key-value pairs
for key, value in student.items():
    print(key, ":", value)
```


## üß™ Lab 9.6: Student Record System (Multiple Dictionaries)


```

# List to store multiple students
students = []

# Collect records for two students
for i in range(2):
    record = {}
    record["name"] = input("Enter name: ")
    record["matric"] = input("Enter matric: ")
    record["score"] = int(input("Enter score: "))
    students.append(record)

# Display all student records
for s in students:
    print(s)
```


## üß™ Lab 9.7: Introduction to Sets

A set stores **unique elements only**.


```

# Create a set with duplicate values
courses = {"CSC701", "CSC702", "CSC701", "CSC703"}

print(courses)  # duplicates removed automatically
```


## üß™ Lab 9.8: Set Operations


```

A = {1, 2, 3}
B = {3, 4, 5}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
```


## üß† Exam Insight: Data Structure Comparison

| Structure | Stores | Ordered | Duplicates |
|---------|-------|--------|-----------|
| List | Values | Yes | Yes |
| Tuple | Values | Yes | Yes |
| Dictionary | Key‚ÄìValue | No | Keys unique |
| Set | Values | No | No |
