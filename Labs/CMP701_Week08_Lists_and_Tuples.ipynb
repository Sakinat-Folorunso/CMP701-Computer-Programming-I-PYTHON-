## üß™ Lab 8.1: Creating a List

A list is an ordered collection of elements enclosed in square brackets `[]`.


```python

# Creating a list of student scores
scores = [65, 70, 55, 80, 75]

# Display the list
print(scores)
```


## üß™ Lab 8.2: List Indexing

Each element in a list has an index starting from 0.


```python

# Accessing elements using index
print(scores[0])   # First element
print(scores[2])   # Third element
print(scores[-1])  # Last element
```


## üß™ Lab 8.3: List Slicing

Slicing allows you to extract a portion of a list.


```python

# Slicing examples
print(scores[1:4])   # Elements from index 1 to 3
print(scores[:3])    # First three elements
print(scores[2:])    # From index 2 to end
```


## üß™ Lab 8.4: Common List Methods


```python

# Append adds an element to the end
scores.append(90)

# Insert adds element at specific index
scores.insert(2, 60)

# Remove deletes first occurrence of value
scores.remove(55)

# Pop removes element at index
scores.pop(1)

# Display modified list
print(scores)
```


## üß† Memory View (Think Like the Computer)

## üß™ Lab 8.5: Processing List Data (Student Scores)


```python

# Create an empty list
scores = []

# Collect five scores from the user
for i in range(5):
    score = int(input("Enter score: "))
    scores.append(score)

# Compute statistics
print("Scores:", scores)
print("Highest score:", max(scores))
print("Lowest score:", min(scores))
print("Average score:", sum(scores)/len(scores))
```


## üß™ Lab 8.6: Introduction to Tuples

A tuple is similar to a list but **cannot be changed** (immutable).


```python

# Creating a tuple for student biodata
student = ("Aisha", "CSC/PGD/001", 78)

# Accessing tuple elements
print(student[0])
print(student[2])
```


## üß† Lists vs Tuples (Exam Favourite)

| Feature | List | Tuple |
|------|------|-------|
| Brackets | [ ] | ( ) |
| Mutable | Yes | No |
| Use case | Changing data | Fixed data |
