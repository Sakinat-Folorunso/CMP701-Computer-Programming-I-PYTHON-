## üß† Concept Recap: What is a String?

A **string** is a sequence of characters enclosed in quotation marks.
Strings are used to store and process text data such as:
- Names
- Sentences
- Passwords
- Documents

In Python, strings are **immutable**.



## üß™ Lab 11.1: Creating and Displaying Strings


```

# Creating string variables
name = "Aisha"
course = "Computer Programming"

# Display strings
print(name)
print(course)
```


## üß™ Lab 11.2: String Indexing

Each character in a string has an index starting from 0.


```

text = "PYTHON"

# Access characters using index
print(text[0])   # First character
print(text[3])   # Fourth character
print(text[-1])  # Last character
```


## üß™ Lab 11.3: String Slicing

Slicing extracts part of a string.


```

word = "COMPUTER"

print(word[0:4])   # COMP
print(word[3:])    # PUTER
print(word[:5])    # COMPU
```


## üß™ Lab 11.4: String Immutability (Important)


```

text = "Python"

# The following line will cause an error if uncommented
# text[0] = 'J'

print(text)
```


## üß™ Lab 11.5: Common String Methods


```

sentence = "  Welcome to CMP701 Programming  "

print(sentence.lower())     # Convert to lowercase
print(sentence.upper())     # Convert to uppercase
print(sentence.strip())     # Remove leading and trailing spaces
print(sentence.replace("Programming", "Python"))
print(len(sentence))        # Length of string
```


## üß™ Lab 11.6: Splitting Strings


```

sentence = "Data Science and Artificial Intelligence"

# Split sentence into words
words = sentence.split()

print(words)
print("Number of words:", len(words))
```


## üß™ Lab 11.7: Word Count Program


```

sentence = input("Enter a sentence: ")

# Split sentence into words
words = sentence.split()

# Display word count
print("Word count:", len(words))
```


## üß™ Lab 11.8: Simple Password Validation


```

password = input("Enter password: ")

if len(password) >= 8:
    print("Password length is acceptable")
else:
    print("Password too short")
```
