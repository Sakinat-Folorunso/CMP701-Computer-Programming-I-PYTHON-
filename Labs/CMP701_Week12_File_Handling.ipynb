## üß† Concept Recap: Why File Handling?

So far, our programs store data **temporarily in memory**.
Once the program ends, all data is lost.

Real-world applications require **permanent storage**, such as:
- Student records
- Examination results
- Transaction logs
- Reports

File handling allows programs to **store and retrieve data from disk**.



## üß™ Lab 12.1: Writing to a Text File

We use the `open()` function to create and write to files.


```

# Open a file in write mode
file = open("students.txt", "w")

# Write data to the file
file.write("Aisha, 78\n")
file.write("Bola, 85\n")
file.write("Chinedu, 90\n")

# Close the file
file.close()

print("Data written successfully")
```


## üß™ Lab 12.2: Reading from a Text File


```

# Open the file in read mode
file = open("students.txt", "r")

# Read entire content
content = file.read()

# Close the file
file.close()

print(content)
```


## üß™ Lab 12.3: Reading Line by Line


```

file = open("students.txt", "r")

for line in file:
    print(line.strip())  # strip removes newline characters

file.close()
```


## üß™ Lab 12.4: Using the `with` Statement (Best Practice)

The `with` statement automatically closes the file.


```

with open("students.txt", "r") as file:
    for line in file:
        print(line.strip())
```


## üß™ Lab 12.5: Appending to a File


```

with open("students.txt", "a") as file:
    file.write("Zainab, 88\n")

print("New record added")
```


## üß™ Lab 12.6: Handling File Errors


```

try:
    with open("nonexistent.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found. Please check the filename.")
```


## üß™ Lab 12.7: Simple File-Based Student Record System


```

# Add new student record
with open("records.txt", "a") as file:
    name = input("Enter student name: ")
    score = input("Enter score: ")
    file.write(name + "," + score + "\n")

# Display all records
with open("records.txt", "r") as file:
    print("Student Records:")
    for line in file:
        print(line.strip())
```
