## üß† Concept Recap: Why Functions?

As programs grow larger, writing all code in one place becomes:
- Hard to read
- Difficult to debug
- Impossible to reuse efficiently

Functions allow us to **break a program into smaller, reusable blocks**.



## üß™ Lab 10.1: Defining and Calling a Function


```

# Define a simple function
def greet():
    # This function prints a welcome message
    print("Welcome to CMP701 Programming")

# Call the function
greet()
```


## üß™ Lab 10.2: Functions with Parameters


```

# Function with parameters
def greet_student(name):
    # Print a personalized greeting
    print("Hello", name)

# Function calls
greet_student("Aisha")
greet_student("Bola")
```


## üß™ Lab 10.3: Functions with Return Values


```

# Function that returns a value
def add(a, b):
    # Compute and return sum
    return a + b

# Store returned value
result = add(5, 3)
print("Sum =", result)
```


## üß† Call-and-Return Flow

Main Program ‚Üí Function Call ‚Üí Execute ‚Üí Return Value ‚Üí Main Continues



## üß™ Lab 10.4: Local and Global Variables


```

# Global variable
x = 10

def show():
    # Local variable
    y = 5
    print("Inside function:", x, y)

show()
print("Outside function:", x)
```


## üß™ Lab 10.5: Modular Calculator Using Functions


```

# Define functions for arithmetic operations

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

# Test the functions
print(add(4, 2))
print(subtract(4, 2))
print(multiply(4, 2))
print(divide(4, 2))
```


## üß™ Lab 10.6: Menu-Driven Program


```

# Menu-driven calculator

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

while True:
    print("1. Add")
    print("2. Subtract")
    print("3. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        x = int(input("Enter first number: "))
        y = int(input("Enter second number: "))
        print("Result:", add(x, y))
    elif choice == "2":
        x = int(input("Enter first number: "))
        y = int(input("Enter second number: "))
        print("Result:", subtract(x, y))
    elif choice == "3":
        break
    else:
        print("Invalid choice")
```
