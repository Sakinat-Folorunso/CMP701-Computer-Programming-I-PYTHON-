## üß† Short Concept Recap (From Lecture)

In programming, **variables store data** that programs work with.

Python is dynamically typed, meaning:
- You do not declare variable types explicitly
- Python determines the type automatically

In this lab, we focus on:

## üß™ Lab 3.1: Variables in Python

A variable is a **named memory location** used to store data.


```python

# Assigning values to variables
age = 25              # age stores an integer value
name = "Sakinat"        # name stores a string value
height = 1.65         # height stores a float value
is_student = True     # is_student stores a boolean value

# Display the variable values
print(age)
print(name)
print(height)
print(is_student)
```


### üîç Explanation
- `=` is the assignment operator
- Variables can store different types of data
- Python determines the data type automatically



## üß™ Lab 3.2: Python Data Types
Python supports several built-in data types.


```python

# Integer type
x = 10
print(type(x))

# Float type
y = 3.14
print(type(y))

# String type
z = "Data Science"
print(type(z))

# Boolean type
flag = False
print(type(flag))
```


## üß™ Lab 3.3: Input and Output

The `input()` function collects data from the user.


```python

# Collecting user input
name = input("Enter your name: ")

# Displaying output
print("Welcome,", name)
```


## üß™ Lab 3.4: Type Conversion

All input collected using `input()` is **text (string)** by default.


```python

# Input two numbers as strings
a = input("Enter first number: ")
b = input("Enter second number: ")

# Convert inputs to integers
a = int(a)
b = int(b)

# Perform addition
print("Sum =", a + b)
```

## üß† Dry-Run (Trace) Example

Study the program below and predict the output before running it.


```python

x = 5
y = 2
z = x + y * 3
print(z)
```


### Dry-Run Explanation
- Multiplication happens before addition
- `y * 3 = 6`
